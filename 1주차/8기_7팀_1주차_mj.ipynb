# 2번 문제

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
# 2011년 이후의 자료만 추출
df_recent = df[df['Year']>= 2011]
# 추출한 자료로 pivot table 만들기
df_recent.pivot_table(values = 'Life_Expectancy', index = 'Year', columns = 'Country', aggfunc = 'mean')
```
