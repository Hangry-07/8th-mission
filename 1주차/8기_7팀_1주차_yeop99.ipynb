# 2번

```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
# 필요없는 column 삭제
df = df.drop(["Spending_USD"], axis=1)
df
```

```python
# 2011년 이전 데이터 삭제
df = df.drop(df[df["Year"]<2011].index)
df
```

```python
# pivot_table 이용
pdf1 = pd.pivot_table(df,                            # 피벗할 데이터프레임
                     index = 'Year',                 # 행 위치에 들어갈 열
                     columns = 'Country',            # 열 위치에 들어갈 열
                     values = 'Life_Expectancy',     # 데이터로 사용할 열
                     aggfunc = 'mean')               # 데이터 집계함수
pdf1
```

```python
# groupby 이용
gdf = df.groupby(['Year','Country']).mean().unstack()
gdf
```
