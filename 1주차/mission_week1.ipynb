```python
##Q2

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")

df = df[df.Year > 2010]   ##2011년부터의 데이터만
df_groups = df.groupby(['Year', 'Country'])['Life_Expectancy' ].mean().round(1)   ##년도와 국가를 기준으로 기대수명의 평균값을 계산
df = pd.DataFrame(df_groups)                                                       ## 다시 데이터프레임형태로 만듦
df.pivot_table(index='Year', columns='Country', values = 'Life_Expectancy')           ##피봇테이블을 이용해 표를 재정렬
```
